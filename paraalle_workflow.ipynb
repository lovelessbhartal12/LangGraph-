```python
from langgraph.graph import StateGraph , START ,END
from typing import TypedDict
```

```python
class batsmenstate(TypedDict):
     runs:int 
     balls:int
     fours:int
     sixes:int

     strike_rate:float
     bpb:float
     boundary_percentage:float
     summary:str
```

```python
def calculate_st(state:batsmenstate):
    sr=(state['runs']/state['balls'])*100
    
    return {'strike_rate':sr}
```

```python
def calculate_bpb(state:batsmenstate):
    bpb=state['balls']/(state['fours']+state['sixes'])
    

    return {'bpb':bpb}
```

```python
def calculate_boundary_percentage(state:batsmenstate):
    boundry_per=((state['fours']*4+state['sixes']*6)/state['runs'])*100
    return {'boundary_percentage':boundry_per}
```

```python
def summary(state:batsmenstate):
    summary=f"""
strike-rate={state['strike_rate']} \n
ball_per_boundary={state['bpb']} \n
boundary_percent={state['boundary_percentage']} \n


    
 """
    
    return {'summary':summary}
```

```python
graph=StateGraph(batsmenstate)

graph.add_node('calcuate_sr', calculate_st)
graph.add_node('calcuate_bpb1', calculate_bpb)
graph.add_node('calcuate_boundary_percentage', calculate_boundary_percentage)
graph.add_node('calcuate_summary', summary)
```

```python
graph.add_edge(START ,'calcuate_sr')
graph.add_edge(START ,'calcuate_bpb1')
graph.add_edge(START ,'calcuate_boundary_percentage')

graph.add_edge('calcuate_sr','calcuate_summary')
graph.add_edge('calcuate_bpb1' ,'calcuate_summary')
graph.add_edge('calcuate_boundary_percentage','calcuate_summary')

graph.add_edge('calcuate_summary' ,END)

workflow=graph.compile()
```

```python
inital_State={
    'runs':100,
    'balls':50,
    'fours':6,
    'sixes':4
}

workflow.invoke(inital_State)
```
