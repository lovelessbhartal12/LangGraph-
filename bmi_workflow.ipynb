```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict
```

```python
#define state of langgraph

class BMIstate(TypedDict):
    weight_kg:float
    height_m:float
    bmi:float 
    category:str
```

```python
def calculate_bmi1(state:BMIstate)->BMIstate:
    weight=state["weight_kg"]
    height=state["height_m"]
    bmi=weight/(height**2)
    state['bmi']=round(bmi ,2)

    return state
```

```python
def labeled_bmi1(state: BMIstate) -> BMIstate:
    bmi=state['bmi']

    if bmi <18.5:
        state['category']="underweight"
    elif 18.5<bmi<25:
        state['category']='normal'
    elif 25< bmi <30:
        state['category']="overweighted"
    else:
    
      state["category"]="obsese"
    return state
```

```python
#define graph
graph=StateGraph(BMIstate)
#add the nodes
graph.add_node('calculate_bmi',calculate_bmi1)
graph.add_node('labeled_bmi', labeled_bmi1)
#ADD edges to your graph

graph.add_edge(START, 'calculate_bmi')
graph.add_edge('calculate_bmi', 'labeled_bmi')
graph.add_edge('labeled_bmi', END)

#compile the graph

workflow=graph.compile()

```

```python
#excute the   graph

initial_state= {'weight_kg':80, 'height_m':1.73}
final_state=workflow.invoke(initial_state)
print(final_state)
```
